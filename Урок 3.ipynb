```python
from math import factorial

def combinations(k, n):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
def get_balls(w, W, N, k):
    """
    :param w : количество вытащенных белых шаров
    :param W : общее количество белых шаров
    :param N : общее количество шаров
    :param k : количество попыток

    :return: вероятность достать w белых шаров из N шаров за k попыток
    """
    # print (f'( combinations({w}, {W}) * combinations({k - w}, {N-W}) ) / combinations({k}, {N})')
    return ( combinations(w, W) * combinations(k - w, N-W) ) / combinations(k, N)
```

#### 1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки. ####

```python
sample = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]
mean = sum(sample)/len(sample)
var = sum([x - mean for x in sample])/len(sample)
var_unbiased = sum([x - mean for x in sample])/(len(sample) -1)
std = var ** (1/2)
print(f'Среднее: {mean}, смещённая дисперсия: {var:.6f}, несмещённая дисперсия: {var_unbiased:.6f}, среднеквадратичное отклонение: {std:.6f}')
```

#### 2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые? ####

**Какова вероятность того, что ровно три мяча белые?** 
Событие X = {ровно три мяча белые}. $X=A\cup B\cup C$, где

| Событие | <p style="text-align: center;"> Описание </p>   |
|------|------|
|<p style="text-align: center;"> A </p>| {в первом ящике 0 белых, во втором 3 белых}|
|<p style="text-align: center;"> B </p>| {в первом ящике 1 белый, во втором 2 белых}|
|<p style="text-align: center;"> C </p>| {в первом ящике 2 белых, во втором 1 белый}|

т.к. события несовместны, то $P(X)=P(A) + P(B) + P(C)$
 
$$ P(A) = \frac{C^0_5 \cdot C^2_3}{C^2_8} \cdot \frac{C^3_5 \cdot C^1_7}{C^4_{12}}= 0.0152 $$
$$ P(B) = \frac{C^1_5 \cdot C^1_3}{C^2_8} \cdot \frac{C^2_5 \cdot C^2_7}{C^4_{12}}= 0.2273 $$
$$ P(C) = \frac{C^2_5 \cdot C^0_3}{C^2_8} \cdot \frac{C^1_5 \cdot C^3_7}{C^4_{12}}= 0.1263 $$

$$ P(X)=P(A) + P(B) + P(C) = 0.0152 + 0.2273 + 0.1263 = 0.3687 $$

```python
A_prob = get_balls(0,5,8,2) * get_balls(3,5,12,4)
B_prob = get_balls(1,5,8,2) * get_balls(2,5,12,4)
C_prob = get_balls(2,5,8,2) * get_balls(1,5,12,4)

print(f'Вероятности событий: A: {A_prob:.4f}, B: {B_prob:.4f}, C: {C_prob:.4f}, X: {A_prob + B_prob + C_prob:.4f}')
```

#### 3. На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.####

Т.к. мы не имеем дополнительной информации будем считать, что спортсмены имели равные шансы совершить выстрел. Таким образом имеем полную группу событий $P(B_i) = \frac 1 3$, легко убедиться что $ \sum_{i=1}^{3}{P(B_i)} = 1$, тогда  $P(A) = \sum_{i=1}^{3}{P(B_i)} \cdot P(A\:|\:B_i)$

| Событие | <p style="text-align: center;"> Описание </p>   | Вероятность события A| Значение
|------|------|------|------
|<p style="text-align: center;"> $B_1$ </p>| {выстрел произведён первым спортсменом}| $P(A|B_1$) | 0.9
|<p style="text-align: center;"> $B_2$ </p>| {выстрел произведён вторым спортсменом}| $P(A|B_2$) | 0.8
|<p style="text-align: center;"> $B_3$ </p>| {выстрел произведён третьим спортсменом}| $P(A|B_3$) |0.6
|<p style="text-align: center;"> $A$ </p> | <p style="text-align: center;">{попадание} </p> |$P(A)$ | 0.7666


Вероятность того, что стрелял i-ый спортсмен — при условии, что выстрел был удачным, — будем находить по формуле:

$$P(B_i\:|\:A) = \frac{P(B_i) \cdot P(A\:|\:B_i)}{P(A)}$$

Подставляем:
$$P(B_1\:|\:A) = \frac{1}{3} \cdot \frac{0.9}{0.7666} \approx 0.3913 $$
$$P(B_2\:|\:A) =  \frac{1}{3} \cdot \frac{0.8}{0.7666} \approx 0.3478 $$
$$P(B_3\:|\:A) =  \frac{1}{3} \cdot \frac{0.6}{0.7666} \approx 0.2608 $$

#### 4. В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C? ####

Из условий задачи имеем, что $C = A \cup B$ и, тогда т.к. A и B не совместны, имеем $P(C) = P(A) + P(B)$, 
т.к. это полная группа событий и $P(A) + P(B) + P(C) = 1$ получаем $$ P(A)=P(B)=\frac{1}{4}, P(C)=\frac{1}{2} $$
Пусть X = {студент сдал первую сессию}, тогда $P(X) = {P(A)} \cdot P(X\:|\:A) + {P(B)} \cdot P(X\:|\:B) + {P(C)} \cdot P(X\:|\:C)$


| Событие | <p style="text-align: center;"> Описание </p>   | Вероятность события A| Значение
|------|------|------|------
|<p style="text-align: center;"> $A$ </p>| {студент с факультета A}| $P(X | A$) | 0.8
|<p style="text-align: center;"> $B$ </p>| {студент с факультета B}| $P(X | B$) | 0.9
|<p style="text-align: center;"> $C$ </p>| {студент с факультета C}| $P(X | C$) | 0.7
|<p style="text-align: center;"> $X$ </p> | <p style="text-align: center;">{студент сдал первую сессию} </p> |$P(X)$| 0.775


Вероятность того, что студент сдавший сессию учится на факультете А вычисляется по формуле:

$$P(A\:|\:X) = \frac{P(A) \cdot P(X\:|\:A)}{P(X)}$$ аналогично для других факультетов, подставяя получаем:

$$P(A\:|\:X) = \frac{1}{4} \cdot \frac{0.8}{0.775} = 0.2580 $$
$$P(B\:|\:X) =  \frac{1}{4} \cdot \frac{0.9}{0.775} = 0.2903 $$
$$P(C\:|\:X) =  \frac{1}{2} \cdot \frac{0.7}{0.775} = 0.4516 $$

#### 5. Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей? ####

Т.к. мы не имеем дополнительной информации будем считать, что устройство в равный частях состоит из трёх деталей. Таким образом имеем полную группу событий $P(B_i) = \frac 1 3$, легко убедиться что $ \sum_{i=1}^{3}{P(B_i)} = 1$, тогда  $P(A) = \sum_{i=1}^{3}{P(B_i)} \cdot P(A\:|\:B_i)$

| Событие | <p style="text-align: center;"> Описание </p>   | Вероятность события A| Значение
|------|------|------|------
|<p style="text-align: center;"> $B_1$ </p>| {деталь №1}| $P(A|B_1$) | 0.1
|<p style="text-align: center;"> $B_2$ </p>| {деталь №2}| $P(A|B_2$) | 0.2
|<p style="text-align: center;"> $B_3$ </p>| {деталь №3}| $P(A|B_3$) |0.25
|<p style="text-align: center;"> $A$ </p> | <p style="text-align: center;">{деталь сломалась} </p> |$P(A)$ | 0.1833


Вероятность того, что сломалась i-ая деталь вычислим по формуле:

$$P(B_i\:|\:A) = \frac{P(B_i) \cdot P(A\:|\:B_i)}{P(A)}$$

Подставляем:
$$P(B_1\:|\:A) = \frac{1}{3} \cdot \frac{0.1}{0.1833} \approx 0.1819 $$
$$P(B_2\:|\:A) =  \frac{1}{3} \cdot \frac{0.2}{0.1833} \approx 0.3637 $$
$$P(B_3\:|\:A) =  \frac{1}{3} \cdot \frac{0.25}{0.1833} \approx 0.4546 $$

**Найти вероятность, что из строя выйдут все детали**

Пусть X = {из строя вышли все детали}, $X = B_1 |\:A \cap B_2 |\:A \cap B_3 |\:A$,  тогда т.к. события $B_i |\:A$ попарно независимы получаем $$P(X) = P( B_1 |\:A \cap B_2 |\:A \cap B_3 |\:A) = P(B_1\:|\:A) \cdot P(B_2\:|\:A) \cdot P(B_3\:|\:A) = 0.0301 $$

**Найти вероятность, что из строя выйдут две детали**

Пусть X = {из строя вышли две детали}, $X = (B_1 |\:A \cap B_2 |\:A ) \cup (B_1 |\:A \cap B_3 |\:A ) \cup (B_2 |\:A \cap B_3 |\:A ) $,  тогда т.к. события $B_i |\:A$ попарно независимы, а события $(B_1 |\:A \cap B_2 |\:A ), (B_1 |\:A \cap B_3 |\:A )$ и $(B_2 |\:A \cap B_3 |\:A )$ попарно несовместны, то  $$P(X) = P( (B_1 |\:A \cap B_2 |\:A ) \cup (B_1 |\:A \cap B_3 |\:A ) \cup (B_2 |\:A \cap B_3 |\:A )) = P(B_1 |\:A \cap B_2 |\:A ) + P(B_1 |\:A \cap B_3 |\:A ) + P(B_2 |\:A \cap B_3 |\:A )$$
$$ P(X) = P(B_1\:|\:A) \cdot P(B_2\:|\:A) + P(B_1\:|\:A) \cdot P(B_3\:|\:A) + P(B_2\:|\:A) \cdot P(B_3\:|\:A) $$
$$ P(X) = 0.1819 * 0.3637 + 0.1819*0.4546 +0.3637 * 0.4546 = 0.3142 $$

**Найти вероятность, что из строя выйдет хотя бы одна деталь**

Пусть X = {из строя вышла одна деталь}, тогда $\overline{X}$ = {из строя не вышло ни одной детали} 

$$\overline{X} = \overline{B_1 |\:A} \cap \overline{B_2 |\:A} \cap \overline{B_3 |\:A}$$,  тогда т.к. события $\overline{B_i |\:A}$ попарно независимы получаем $$P(\overline{X}) = (1-0.1819) * (1-0.3637) * (1 - 0.4546) = 0.2839$$, тогда $P(X) = 1 - 0.2839 = 0.7161 $

**Найти вероятность, что из строя выйдет от одной до двух деталей**

Пусть X = {из строя вышла от одной до двух деталей}, тогда 
$\overline{X}$ = {из строя не вышло ни одной детали или из строя вышли все детали},
тогда $P(\overline{X}) = $ P({вышли все детали} + {не вышло ни одной детали} = 0.0301 + 0.7161 = 0.7462

Тогда $P(X) = 1 - 0.7462 = 0.2538 $
