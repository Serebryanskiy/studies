### Доработать приложение по поиску авиабилетов, чтобы оно возвращало билеты по названию города, а не по IATA коду. Пункт отправления и пункт назначения должны передаваться в качестве параметров. Сделать форматированный вывод, который содержит в себе пункт отправления, пункт назначения, дату вылета, цену билета (можно добавить еще другие параметры по желанию) ###

```python
from datetime import datetime
import requests
import json
```

```python
IATA_list = requests.get('http://api.travelpayouts.com/data/en/cities.json').json()
IATA_dict = {city['name']:city['code'] for city in IATA_list}
print(IATA_dict['Moscow'])
```

```python
def get_prices(origin, destination, depart_date=datetime.today().strftime('%Y-%m-%d'), one_way=True):
    params = dict(origin=IATA_dict[origin],destination=IATA_dict[destination], depart_date=depart_date,one_way=one_way)
    request = requests.get("http://min-prices.aviasales.ru/calendar_preload", params=params).json()
    output = [dict(price=deal['value'],date=deal['depart_date'],origin=origin, destination=destination,) for deal in request['best_prices']]
    return output

print(json.dumps(get_prices('Moscow','Novosibirsk'),indent=4))
```

### В приложении парсинга википедии получить первую ссылку на другую страницу и вывести все значимые слова из неё. Результат записать в файл в форматированном виде ###

```python
from requests import get
import re
from bs4 import BeautifulSoup

def get_link(topic):
    link = "https://ru.wikipedia.org/wiki/" + topic.capitalize()
    return link

def get_topic_page(topic):
    link = get_link(topic)
    html = get(link).text
    return html

def get_topic_text(topic):
    html_content = get_topic_page(topic)
    words = re.findall("[а-яА-Я]{3,}",html_content)
    return words

def get_common_words(topic):
    words_list = get_topic_text(topic)
    rate={}
    for word in words_list:
        if word in rate:
            rate[word]+=1
        else:
            rate[word]=1
    rate_list = list(rate.items())
    rate_list.sort(key = lambda x :-x[1])

    return rate_list

def visualize_common_words(topic):
    words = get_common_words(topic)
    for w in words[0:10]:
        print(f'{w[0]} встречается {w[1]} раз')


def save_first_link_words(topic):
    response = get("https://ru.wikipedia.org/wiki/" + topic.capitalize()).content
    soup = BeautifulSoup(response, 'html.parser')
    link = soup.find('body').find('a', attrs={"rel": "nofollow","class": "external text"}).attrs['href']

    html_content = get(link).text
    words_list = re.findall("[a-zA-Zа-яА-ЯёЁ]{3,}",html_content)
    rate={}
    for word in words_list:
        if word in rate:
            rate[word]+=1
        else:
            rate[word]=1
    rate_list = list(rate.items())
    rate_list.sort(key = lambda x :-x[1])
    words = rate_list
    for w in words[0:10]:
        print(f'{w[0]} встречается {w[1]} раз')
    

save_first_link_words('математика')
#list1 = visualize_common_words('математика')
#print(list1)
#print(len(list1))
```
