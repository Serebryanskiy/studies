```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

ax=Axes3D(figure())

x=np.arange(-5,5.5,0.5)
y=np.arange(-5,5.5,0.5)

x,y=np.meshgrid(x,y)
z=np.sqrt(x**2+y**2)
z2=-np.sqrt(x**2+y**2)

ax.plot_surface(x,y,z)
ax.plot_surface(x,y,z2)

```
