```python
# 4. Дополните код расчетом коэффициента корреляции x и y по формуле

import numpy as np
import matplotlib.pyplot as plt

n = 100
r = 0.7
x = np.random.rand(n)
y = r * x + (1 - r) * np.random.rand(n)

plt.plot(x, y, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

a = (np.sum(x) * np.sum(y) - n * np.sum(x * y)) / (np.sum(x) * np.sum(x) - n * np.sum(x * x))
b = (np.sum(y) - a * np.sum(x)) / n
    
A = np.vstack([x, np.ones(len(x))]).T
a1, b1 = np.linalg.lstsq(A, y,rcond = None)[0]

print(a, b)
print(a1, b1)

cov = np.average( (x-np.average(x))*(y-np.average(y)) ) # коэффициент ковариации
dispx = np.average( np.power( (x-np.average(x)), 2) )   # дисперсия х
dispy = np.average( np.power( (y-np.average(y)), 2) )   # дисперсия y 

corr = cov/(np.sqrt(dispx)*np.sqrt(dispy)) # коэффициент корреляции
print({"расчётное":corr, "программное":np.corrcoef(x,y)[0,1]})

plt.plot([0, 1], [b, a + b])
plt.show()
```
