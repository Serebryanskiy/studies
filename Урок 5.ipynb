**1. Известно, что генеральная совокупность распределена нормально со средним квадратическим отклонением, равным 16. Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95, если выборочная средняя M = 80, а объем выборки n = 256.**

Т.к. нормальное распределение устойчиво по суммированию, то если наблюдение $X_i$ имеет нормальное распределение, то велечина 


$\eta = \sqrt{n}\frac{M-a}{\sigma}\ (1) $ имеет стандартное нормальное распределение.
Тогда найдём такое c>0, что $P(-c < \eta < c) = 0.95\ (2)$

Используя свойства функции распределение стандартного нормального закона полуаем, что 
$$ P(-c < \eta < c) = 2\Phi_{0,1}(с) -1 => \Phi_{0,1}(с) = \frac{(0.95 + 1)}{2}=0.975$$
используя табличные значения получаем, что c = 1.96

Теперь подставим (1) в (2) и разрешим неравенство относительно a:
$$P(-1.96 < \eta < 1.96) = P(M-1.96  <  \sqrt{n}\frac{M-a}{\sigma} < 1.96) = P(M - 1.96\frac{\sigma}{\sqrt{n}} < a < M + 1.96\frac{\sigma}{\sqrt{n}})=P(78.04 < a < 81.96) = 0.95 $$
Таким образом мы получили, что доверительный интервал (79.04,81.96) c вероятностью 95% накрывает параметр математитческого ожидания a

**2. В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью, получены опытные данные: 6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1 Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей, оценить истинное значение величины X при помощи доверительного интервала, покрывающего это значение с доверительной вероятностью 0,95.**

Дано: выборка X = [6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1]
Посчитаем среднее и несмещённую дисперсию, M = 6.59, $S_0^2 = 0.2032, S_0=0.4508$
Используя случайную величину $$\eta = \sqrt{n}\frac{M-a}{S_0}$$ которая имеет распределение Стьюдента с n-1 степенью свободы и ту же логику, что и в первом задании получаем, что 
$$ P(-c < \eta < c) = 2F_T^{n-1}(с) -1 => F_{T}^{9}(с) = \frac{(0.95 + 1)}{2}=0.975$$
используя табличные значения получаем, что c = 2.2621, далее

$$P(-2.2621 < \eta < 2.2621) =  P(M - 2.2621\frac{S_0}{\sqrt{n}} < a < M + 1.96\frac{S_0}{\sqrt{n}})=P(6.59 - 0.3225 < a < 6.59 + 0.3225) =  P(6.2675 < a < 6.9125) = 0.95 $$
Таким образом мы получили, что доверительный интервал (6.2675,6.9125) c вероятностью 95% накрывает параметр математитческого ожидания a величины X


```python
sample = [6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1]

mean_ = sum(sample)/len(sample)
var_ = sum([(x - mean_) ** 2 for x in sample]) / len(sample)
var_unbiased = sum([(x - mean_) ** 2 for x in sample]) / (len(sample) - 1)
std_ = var_unbiased ** (1/2)

print(f'Среднее: {mean_:.4f}, смещённая дисперсия: {var_:.4f}, несмещённая дисперсия: {var_unbiased:.4f}, среднеквадратичное отклонение: {std_:.4f}')
print(f'Дельта: {2.2621*std_/(len(sample)**(1/2)):.4f}')
```

**3. Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм. Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр оказался равным 17.5 мм, а дисперсия известна и равна 4 мм.**

H0: M = 17

H1: M != 17


$U_{набл} = \sqrt{n}\frac{M-a_0}{\sigma} = \sqrt{100}\frac{17.5-17}{2} = 2.5$

Построим критерий размера 0.05, $\tau_{0.95} = 1.645$:
$$ \delta =
  \begin{cases}
    H0      & U_{набл} < \tau_{0.95}\\
    H1      & U_{набл} \ge \tau_{0.95}
  \end{cases}
$$

Т.к. $U_{набл} \ge \tau_{0.95} (2.5 > 1.645)$ то принимаем гипотезу H1 и отвергаем гипотезу H0 о среднем диаметре в 17мм 

**4. Продавец утверждает, что средний вес пачки печенья составляет 200 г. Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет: 202, 203, 199, 197, 195, 201, 200, 204, 194, 190. Известно, что их веса распределены нормально. Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%?**

Дано: выборка X = [202, 203, 199, 197, 195, 201, 200, 204, 194, 190]
Посчитаем среднее и несмещённую дисперсию, M = 198.5, $S_0^2 = 19.8333, S_0=4.4535$
Используя случайную величину $$\eta = \sqrt{n}\frac{M-a}{S_0}$$ которая имеет распределение Стьюдента с n-1 степенью свободы и ту же логику, что и в первом задании получаем, что 
$$ P(-c < \eta < c) = 2F_T^{n-1}(с) -1 => F_{T}^{9}(с) = \frac{(0.99 + 1)}{2}=0.995$$
используя табличные значения получаем, что c = 3.2498, далее

$$P(198.5 - 4.5767 < a < 198.5 + 4.5767) =  P(193.9233 < a < 203.0767) = 0.99 $$
Таким образом мы получили, что доверительный интервал (193.9233,203.0767) c вероятностью 99% накрывает параметр математитческого ожидания a величины. Утверждение продавца верно, т.к. его оценка попадает в ДИ.

```python
sample = [202, 203, 199, 197, 195, 201, 200, 204, 194, 190] 

mean_ = sum(sample)/len(sample)
var_ = sum([(x - mean_) ** 2 for x in sample]) / len(sample)
var_unbiased = sum([(x - mean_) ** 2 for x in sample]) / (len(sample) - 1)
std_ = var_unbiased ** (1/2)

print(f'Среднее: {mean_:.4f}, смещённая дисперсия: {var_:.4f}, несмещённая дисперсия: {var_unbiased:.4f}, среднеквадратичное отклонение: {std_:.4f}')
print(f'Дельта: {3.2498*std_/(len(sample)**(1/2)):.4f}')
```
