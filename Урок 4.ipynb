#### 1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800]. Найдите ее среднее значение и дисперсию.###

В нашем случае, a = 200, b = 800, тогда имеет следующую функцию распределения:
$$f(x) = \begin{cases} 0, если \; x \le 200; \\ \frac{1}{600}, если \; 200 < x \le 800; \\ 0, \; если \; x > 800. \end{cases}$$
Тогда среднее: $$M(A) = \frac{a + b}{2}=\frac{200 + 800}{2}=500$$
Дисперсия: $$D(B) = \frac{(b - a)^2}{12} = \frac{(800 - 200)^2}{12} = 30 000$$


#### 2. О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2. Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5? Если да, найдите ее. ###

Из условий задачи имеем $a=0.5,\ D(B)=\frac{(b - a)^2}{12}=0.2$ 
тогда $$b-a = \pm\sqrt{0.2 \cdot 12},\ т.к.\ b\ не\ может\ быть\ меньше\ a,\ то$$
$$b=\sqrt{0.2 \cdot 12} + 0.5 = 2.049$$
$$M(B) = \frac{a + b}{2} = \frac{0.5 + 2.049}{2} = 1.2745 $$

#### 3. Непрерывная случайная величина C распределена нормально и задана плотностью распределения $f(x) = \frac{1}{4 \sqrt{2 \pi}} e ^ {- \frac{(x + 2)^2}{32}}$. Найдите: а). M(X) б). D(X) в). std(X) (среднее квадратичное отклонение)

Нормальное распределение $(N_{a,\sigma^2}$) в общем виде с параметрами $a$ и $\sigma^2$ имеет вид: 
$$f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e ^ {- \frac{(x - a)^2}{2 \sigma^2}},$$
тогда очевидно, что случайная велечина $C \in N_{-2,16}$, где $a=-2$ и $\sigma^2 = 16$.
Тогда $M(C)=-2, D(C)=16, std(C) = 4$


#### 4. Рост взрослого населения города X имеет нормальное распределение. Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см. Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост: а). больше 182 см б). больше 190 см в). от 166 см до 190 см г). от 166 см до 182 см д). от 158 см до 190 см е). не выше 150 см или не ниже 190 см ё). не выше 150 см или не ниже 198 см ж). ниже 166 см. ###
    
    

Пусть $\xi$ - это случайная велечина принимающая значения роста взрослого населения города X, тогда из условий задачи, $\xi \in N_{174,64}$ и имеет плотность распределения: $$f(x) = \frac{1}{8 \sqrt{2 \pi}} e ^ {- \frac{(x - 174)^2}{128}}$$
Для того, чтобы ответить на вопросы задания, вспомним некоторые свойства функции нормального распределения:
$$ \fbox{$Для\ любого\ x \in R\ справедливо\ соотношение:
\Phi_{a,\sigma^2}(x) = \Phi_{0,1}(\frac {x-a}{\sigma})$}\ (1) $$
$$ \fbox{$Если\ \xi \in N_{a,\sigma^2},\ то:
P(x_1 < \xi < x_2) = \Phi_{a,\sigma^2}(x_2) - \Phi_{a,\sigma^2}(x_1) = \Phi_{0,1}(\frac {x_2-a}{\sigma}) - \Phi_{0,1}(\frac {x_1-a}{\sigma})$}\ (2) $$
$$ \fbox{$
P(\xi > x) = 1 - P(\xi \le x) =1 - P(\xi \le x) - P(\xi = x) = 1 - \Phi_{a,\sigma^2}(x)$}\ (3) $$
$$ \fbox{$\Phi_{0,1}(-x) = 1 - \Phi_{0,1}(x)$}\ (4) $$

**a) больше 182 см**


Используем свойства (1) и (3)
$$P(\xi > 182) = 1 - P(\xi \le 182) = 1 - \Phi_{174,64}(182) = 1 - \Phi_{0,1}(\frac {182-174}{8}) = 1 - \Phi_{0,1}(1) = 0.15865$$
**б) больше 190 см**


Используем свойство (1) и (3) 
$$P(\xi > 190) = 1 - P(\xi \le 190) = 1 - \Phi_{174,64}(190) = 1 - \Phi_{0,1}(\frac {190-174}{8}) = 1 - \Phi_{0,1}(2) = 0.02275$$
**в) от 166 см до 190 см**


Используем свойства (1) и (2) и (4)

$$P(166 <\xi < 190) = \Phi_{0,1}(\frac {190-174}{8}) - \Phi_{0,1}(\frac {166-174}{8}) = \Phi_{0,1}(2) - \Phi_{0,1}(-1) = \Phi_{0,1}(2) - (1 - \Phi_{0,1}(1)) = 0.81859$$

**г) от 166 см до 182 см**


Используем свойства (1) и (2) и (4)

$$P(166 <\xi < 182) = \Phi_{0,1}(\frac {182-174}{8}) - \Phi_{0,1}(\frac {166-174}{8}) = \Phi_{0,1}(1) - \Phi_{0,1}(-1) = \Phi_{0,1}(1) - 1 + \Phi_{0,1}(1)) = 2\Phi_{0,1}(1) - 1 = 0.68268$$
**д) от 158 см до 190 см**


Используем свойства (1) и (2) и (4)


$$P(158 <\xi < 190) = \Phi_{0,1}(\frac {190-174}{8}) - \Phi_{0,1}(\frac {158-174}{8}) = \Phi_{0,1}(2) - \Phi_{0,1}(-2) = \Phi_{0,1}(2) - 1 + \Phi_{0,1}(2)) = 2\Phi_{0,1}(2) - 1 = 0.95449$$
**е) не выше 150 см или не ниже 190 см**


 Пусть событие $ A = \{\xi < 150\} \cup \{\xi > 190\} $
Тогда учитывая что эти события несовместны используя свойство аддитивности вероятности получаем:
$$ P(A) = P (\{\xi < 150\} \cup \{\xi > 190\}) = P (\xi < 150) + P(\xi > 190)  = \Phi_{0,1}(\frac{150-174}{8}) + 1 - \Phi_{0,1}(2) = 1 - \Phi_{0,1}(3) + 1 - \Phi_{0,1}(2) = 0.02410$$
**ё) не выше 150 см или не ниже 198 см**


 Пусть событие $ A = \{\xi < 150\} \cup \{\xi > 198\} $
Тогда учитывая что эти события несовместны используя свойство аддитивности вероятности получаем:
$$ P(A) = P (\{\xi < 150\} \cup \{\xi > 198\}) = P (\xi < 150) + P(\xi > 198)  = \Phi_{0,1}(\frac{150-174}{8}) + 1 - \Phi_{0,1}(3) = 2 - 2\Phi_{0,1}(3) = 0.00269$$

**ё) ниже 166**

$$P(\xi < 166) = \Phi_{174,64}(166) = \Phi_{0,1}(\frac {166-174}{8}) = \Phi_{0,1}(-1) = 1 - \Phi_{0,1}(1) = 0.15865$$

#### 5. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см? ####
    
  Отклонение от математического ожидания равно $190-178 = 12, \sigma = 5$, следовательно рост человека отклоняется на 12/5 сигм, т.е. на $2.4\sigma$
