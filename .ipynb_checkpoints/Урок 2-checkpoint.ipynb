### 1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз. ###

Вероятность события без учёта возможных перестановок равна: $0.8 ^{85} \cdot 0.2^{15}$ , помня, что есть ровно $C_{100}^{85}$ способов расположить  85 успехов в 100 местах получаем, что вероятность попасть в цель равна: $C_{100}^{85}\cdot 0.8 ^{85} \cdot 0.2^{15} = 0.048061793700746556 $

### 2. Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две? ###

Вероятность того, что ни одна лампочка не перегорит равна $0.0004 ^{0} \cdot 0.9996^{5000} = 0.1352811455$

Вероятность того, что перегорят ровно две равна $C_{5000}^2 \cdot 0.0004 ^{2} \cdot 0.9996^{4998}$, применим формулу Пуассона для приблизительного вычисления вероятности.

По условиям задачи $m = 2$ и $λ = np = 5000 \cdot 0.0004 = 2$:

$$P_{2} \approx \frac{\lambda^m}{m!}e^{-\lambda} \approx \frac{2^{2}}{2!}e^{-2} \approx  0.2706705664732254$$

### 3. Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз? ###

Вероятность события без учёта возможных перестановок равна: $0.5 ^{70} \cdot 0.5^{74}$ , помня, что есть ровно $C_{144}^{70}$ способов расположить  70 успехов в 144 местах получаем, что вероятность выпадения 70 орлов из 144 экспериментов равна: $$C_{144}^{70}\cdot 0.5 ^{70} \cdot 0.5^{74} = 0.06281178035144776 $$

### 4. В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые? Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый? ###

**Какова вероятность того, что все мячи белые?** 
Вероятность того, что все мячи белые равна вероятности пересечания события А = {из первой корзины достали два белых мяча} и события B = {из второй корзины достали два белых мяча}. Т.к. эти события независимы получаем: P($A\cap B) = P(A) \cdot P(B)$

$$P(A) = \frac{C_7^2}{C_{10}^2}= 0.4666,P(B) = \frac{C_9^2}{C_{11}^2}= 0.6545,  P(A\cap B) = 0.4666 \cdot 0.6545 =0.3054$$

**Какова вероятность того, что ровно два мяча белые?** 
Событие X = {ровно два меча белые}. $X=A\cup B\cup C$, где A = {в первой корзине 0 белых, во второй 2 белых}, B = {в первой корзине 2 белых, во второй 0 белых}, C = {в первой корзине 1 белый, во второй 1 белый}, т.к. события несовместны, то
$P(X)=P(A) + P(B) + P(C)$
 
$$ P(A) = \frac{C_3^2}{C_{10}^2} \cdot \frac{C_9^2}{C_{11}^2}= 0.0666 \cdot 0.6545 = 0.0435 $$
$$ P(B) = \frac{C_7^2}{C_{10}^2} \cdot \frac{C_2^2}{C_{11}^2}= 0.4666 \cdot 0.0181 = 0.0084 $$
$$ P(C) = \frac{C_7^1 \cdot C_3^1}{C_{10}^2} \cdot \frac{C_9^1 \cdot C_2^1}{C_{11}^2}= 0.4666 \cdot 0.3272 = 0.1527 $$
 
$$ P(X)=P(A) + P(B) + P(C) = 0.0435 + 0.0084 + 0.1527 = 0.2046 $$

**Какова вероятность того, что хотя бы один мяч белый?** 

Событие A = {хотя бы один мяч белый} P(A) = 1 - P(B), где B = {ни одного белого}


$$ P(B) = \frac{C_3^2}{C_{10}^2} \cdot \frac{C_2^2}{C_{11}^2}= 0.0666 \cdot 0.0181 = 0.0012 $$

Тогда P(A) = 1 - 0.0012 = 0.9988
