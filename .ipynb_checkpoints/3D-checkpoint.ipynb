```python
%matplotlib inline
import nunmpy as np
import matplotlib.pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

x=[];y=[];minusy=[]

def circle(x0,y0,R):
    for i in range(-1000,1000):
        _x=i/100
        underroot=R ** 2 - (_x - x0)** 2
        if underroot>=0: 
            _y=math.sqrt(underroot)
            x.append(_x)
            y.append(_y)
            minusy.append(-_y)
    plt.plot(x,y)
    plt.plot(x,minusy)

circle(0,0,3)
```
