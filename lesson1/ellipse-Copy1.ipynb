```python
%matplotlib inline
import matplotlib.pyplot as plt
import math

# (x - x0)** 2 / a ** 2 + (y -y0)** 2 / b ** 2 = 1      уравнение эллипса проходящего
# ( (y -y0)** 2 / b ** 2 = 1 - (x - x0)** 2 / a ** 2    через точки x0 и y0
# ( (y -y0)** 2 / b ** 2 = b**2* (1 - (x - x0)** 2 / a ** 2) 
# y = sqrt(b**2* (1 - (x - x0)** 2 / a ** 2) ) + y0

x=[];y=[];minusy=[]

def ellipse(x0,y0,a,b):
    for i in range(-1000,1000):
        _x=i/100
        underroot=b ** 2 * (1- (_x-x0) ** 2 / a ** 2)
        if underroot>=0: 
            _y=math.sqrt(underroot)+y0
            x.append(_x)
            y.append(_y)
            minusy.append(-_y)
    plt.plot(x,y)
    plt.plot(x,minusy)

elipse(0,0,3,2)
```
