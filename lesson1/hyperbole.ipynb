```python
%matplotlib inline
import matplotlib.pyplot as plt
import math

# (x - x0)** 2 / a ** 2 - (y -y0)** 2 / b ** 2 = 1      уравнение гиперболы проходящего
# ( (y -y0)** 2 / b ** 2 = (x - x0)** 2 / a ** 2 - 1   через точки x0 и y0
# ( (y -y0)** 2 / b ** 2 = b**2* ((x - x0)** 2 / a ** 2 - 1) 
# y = sqrt(b**2* ((x - x0)** 2 / a ** 2) -1) + y0

x=[];y=[];minusy=[]

def hyper(x0,y0,a,b):
    for i in range(-500,501):
        _x=i/100
        underroot=b ** 2 * ((_x-x0) ** 2 / a ** 2 - 1)
        if underroot>=0: 
            _y=math.sqrt(underroot)+y0
            x.append(_x)
            y.append(_y)
            minusy.append(-_y)
    plt.plot(x,y)
    plt.plot(x,minusy)

hyper(0,0,1,1)
```
